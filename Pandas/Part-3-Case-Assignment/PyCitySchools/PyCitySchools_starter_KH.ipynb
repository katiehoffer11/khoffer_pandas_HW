```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of students

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

```python
#Calculate the total number of schools
number_of_schools = school_data_complete["school_name"].nunique()
number_of_schools
```

```python
#Calculate the total number of students
number_of_students = school_data["size"].sum()
number_of_students
```

```python
#Calculate the total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
#Calculate the average math score
avg_math_score = student_data["math_score"].mean()
avg_math_score
```

```python
#Calculate the average reading score
avg_reading_score = student_data["reading_score"].mean()
avg_reading_score
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
math_70_plus = student_data.loc[student_data["math_score"] >= 70, :]
math_70_plus_count = len(math_70_plus)

percent_students_math_70_plus = (math_70_plus_count / number_of_students)*100
percent_students_math_70_plus
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
reading_70_plus = student_data.loc[student_data["reading_score"] >= 70, :]
reading_70_plus_count = len(reading_70_plus)

percent_students_reading_70_plus = (reading_70_plus_count / number_of_students)*100
percent_students_reading_70_plus
```

```python
#Calculate the percentage of students who passed (70%+) both math and reading (% Overall Passing)

math_reading_70_plus = student_data.loc[(student_data["reading_score"] >= 70) & (student_data["math_score"] >= 70), :]
math_reading_70_plus_count = len(math_reading_70_plus)

percent_students_math_reading_70_plus = (math_reading_70_plus_count / number_of_students)*100
percent_students_math_reading_70_plus
```

```python
#Create a dataframe to hold the above results

new_school_data = pd.DataFrame({"Total Number of Schools":[number_of_schools],
                                "Total Number of Students":[number_of_students],
                                "Total Budget":[total_budget],
                                "Avg Math Score":[avg_math_score],
                                "Avg Reading Score":[avg_reading_score],
                                "%Passing Math":[percent_students_math_70_plus],
                                "%Passing Reading":[percent_students_reading_70_plus],
                                "%Passing Math and Reading":[percent_students_math_reading_70_plus]})

#Optional: give the displayed data cleaner formatting
new_school_data["Total Number of Students"] = new_school_data["Total Number of Students"].map("{:,}".format)
new_school_data["Total Budget"] = new_school_data["Total Budget"].map("${:,}".format)
new_school_data["Avg Math Score"] = new_school_data["Avg Math Score"].map("{:.0f}".format)
new_school_data["Avg Reading Score"] = new_school_data["Avg Reading Score"].map("{:.0f}".format)
new_school_data["%Passing Math"] = new_school_data["%Passing Math"].map("{:.0f}%".format)
new_school_data["%Passing Reading"] = new_school_data["%Passing Reading"].map("{:.0f}%".format)
new_school_data["%Passing Math and Reading"] = new_school_data["%Passing Math and Reading"].map("{:.0f}%".format)

new_school_data
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#need to change per student budget back to int for bins
#school_grouped_df["Per Student Budget"] = pd.to_numeric(school_grouped_df["Per Student Budget"])

bins = [0, 585, 615, 645, 700]
group_names = ["<$585", "$585-$615", "$615-$645", "$645+"]

school_grouped_df["Spending Range"] = pd.cut(school_grouped_df["Per Student Budget"], bins, labels=group_names, include_lowest=True)

grouped_by_spending_range = school_grouped_df.groupby(["Spending Range"])

math_by_spending_range = grouped_by_spending_range["Avg Math Score"].mean()
reading_by_spending_range = grouped_by_spending_range["Avg Reading Score"].mean()
percent_math_by_spending_range = grouped_by_spending_range["%Passing Math"].mean()
percent_reading_by_spending_range = grouped_by_spending_range["%Passing Reading"].mean()
percent_overall_by_spending_range = grouped_by_spending_range["%Overall Passing Rate"].mean()

scores_by_school_spending = pd.DataFrame({
    'Avg Math Score':math_by_spending_range, 
    'Avg Reading Score':reading_by_spending_range,
    '% Passing Math':percent_math_by_spending_range,
    '%Passing Reading':percent_reading_by_spending_range,
    '%Overall Passing Rate':percent_overall_by_spending_range})
scores_by_school_spending
```
