```python
# Dependencies
import pandas as pd
```

```python
# load CSV
path = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
# You may need to encode this file
new_coders_pd = pd.read_csv(path)
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111

new_coders_pd = new_coders_pd.iloc[:,[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
new_coders_pd.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
fixes = {
    0:'No',
    1:'Yes'
}

new_coders_pd = new_coders_pd.replace(fixes)
new_coders_pd.head()
```

```python
# Extract rows for only those who attended a bootcamp
attended_bootcamp = new_coders_pd.loc[new_coders_pd["AttendedBootcamp"] == "Yes", :]
attended_bootcamp.head()
```

```python
# Create DataFrame of the various boot camps along with the number of participants (enrollees)
# Create a DataFrame from using value_counts
bootcamp_values = attended_bootcamp["BootcampName"].value_counts()

Bootcamp_counts_df = pd.DataFrame(bootcamp_values)
Bootcamp_counts_df = Bootcamp_counts_df.reset_index()
Bootcamp_counts_df.columns = ["Bootcamp Name", "counts"]

Bootcamp_counts_df.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
attendess_recommend = attended_bootcamp.loc[attended_bootcamp["BootcampRecommend"] == "Yes", :]

attendess_recommend_grouped = attendess_recommend.groupby("BootcampName")
bootcamp_recommend_values = attendess_recommend_grouped["BootcampRecommend"].count()


bootcamp_recommend_df = pd.DataFrame(bootcamp_recommend_values)
bootcamp_recommend_df = bootcamp_recommend_df.reset_index()
bootcamp_recommend_df.columns = ["Bootcamp Name", "Recommenders"]

bootcamp_recommend_df.head()
```

```python
# Merge the two created data frames on the name of tbe bootcamp

merge_df = pd.merge(Bootcamp_counts_df, bootcamp_recommend_df, on="Bootcamp Name", how="outer")
merge_df
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
merge_df["% Reccommend"] = (merge_df["Recommenders"]/ merge_df["counts"])*100

# Sort results in descending order
sorted_merged_df = merge_df.sort_values(["% Reccommend"], ascending=False)

# Format for percentages
sorted_merged_df["% Reccommend"] = sorted_merged_df["% Reccommend"].map("{:.2f}%".format)
sorted_merged_df
```

```python
# Export to excel and remove index
sorted_merged_df.to_excel("output/Bootcamp2_KH.xlsx", index=False)
```
